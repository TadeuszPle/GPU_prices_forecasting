## Forecasting

```python
train=df_NVidia.iloc[:-30]
test=df_NVidia.iloc[-30:]
start=len(train)
end=start+len(test)-1
df_NVidia
```

```python
result = seasonal_decompose(df_NVidia.iloc[-90:], model='mul')
result.seasonal.plot(figsize=(16,6));
```

There is apparent seasonality, but no trend. 

### Holt's Winter

```python
fit_add_add = ExponentialSmoothing(train, trend="add", seasonal="add", seasonal_periods=7).fit()
fit_add_mul = ExponentialSmoothing(train, trend="add", seasonal="mul", seasonal_periods=7).fit()
fit_mul_add = ExponentialSmoothing(train, trend='mul', seasonal='add', seasonal_periods=7).fit()
fit_mul_mul = ExponentialSmoothing(train, trend="mul", seasonal="mul", seasonal_periods=7).fit()
fit_none_add = ExponentialSmoothing(train, trend=None, seasonal='add', seasonal_periods=7).fit()
fit_none_mul = ExponentialSmoothing(train, trend=None, seasonal="mul", seasonal_periods=7).fit()


fcast_add_add = fit_add_add.forecast(len(test)).rename("additive trend, additive seasonal")
fcast_add_mul = fit_add_mul.forecast(len(test)).rename("additive trend, multiplicative seasonal")
fcast_mul_add = fit_mul_add.forecast(len(test)).rename("multiplicative trend, additive seasonal")
fcast_mul_mul = fit_mul_mul.forecast(len(test)).rename("multiplicative trend, multiplicative seasonal")
fcast_none_add = fit_none_add.forecast(len(test)).rename("none trend, additive seasonal")
fcast_none_mul = fit_none_mul.forecast(len(test)).rename("none trend, multiplicative seasonal")

print(f"additive trend, additive seasonal \t MAPE={mean_absolute_percentage_error(test['Price_USD'], fcast_add_add):.4f}")
print(f"additive trend, multiplictive seasonal \t MAPE={mean_absolute_percentage_error(test['Price_USD'], fcast_add_mul):.4f}")
print(f"multiplictive trend, additive seasonal \t MAPE={mean_absolute_percentage_error(test['Price_USD'], fcast_mul_add):.4f}")
print(f"multiplictive trend, multiplictive seasonal \t MAPE={mean_absolute_percentage_error(test['Price_USD'], fcast_mul_mul):.4f}")
print(f"none trend, additive seasonal \t MAPE={mean_absolute_percentage_error(test['Price_USD'], fcast_none_add):.4f}")
print(f"none trend, multiplictive seasonal \t MAPE={mean_absolute_percentage_error(test['Price_USD'], fcast_none_mul):.4f}")
```

```python
fitHoltWinter=ExponentialSmoothing(train['Price_USD'], trend=None, seasonal='add', seasonal_periods=7).fit()
fcastHoltWinter=fitHoltWinter.forecast(len(test)).rename('Holt-Winter\'s')
mean_absolute_percentage_error(fcastHoltWinter, test['Price_USD'])
```

```python
df_caus=pd.DataFrame([])
df_caus['NVidia']=df_NVidia['Price_USD']
df_caus['Eth']=df_Crypto
grangercausalitytests(df_caus[['NVidia', 'Eth']], maxlag=5)
```

### Vector autoregressive (VAR)

```python
trainVar=df_caus.iloc[:-30]
testVar=df_caus.iloc[-30:]
minAIC=np.inf
minAIC_ord=0
for i in range(1, 50):
    modelVAR=VAR(trainVar)
    resultsVAR=modelVAR.fit(i)
    if resultsVAR.aic<minAIC:
        minAIC=resultsVAR.aic
        minAICord=i
    #print(f'order = {i}; \t AIC = {resultsVAR.aic}')
print(f'best order = {minAICord}; \t AIC = {minAIC}')
```

```python
modelVAR=VAR(trainVar)
resultsVAR=modelVAR.fit(33)
resultsVAR.summary()
lag_order=resultsVAR.k_ar
predictionsVar=resultsVAR.forecast(trainVar.values[-lag_order:],30)
testValuesNV=testVar['NVidia'].values
testValuesEth=testVar['Eth'].values
fcastVAR=pd.DataFrame(predictionsVar[:,0])
fcastVAR.set_index(test.index, inplace=True)
fcastVAR.columns=['NVidia']
mean_absolute_percentage_error(predictionsVar[:,0], testValuesNV)
```

### Prophet

```python
model=Prophet(yearly_seasonality=False, weekly_seasonality=True)
data=train.reset_index()
data.columns=['ds', 'y']
model.fit(data)
future=model.make_future_dataframe(periods=len(test), freq='D')
forecast=model.predict(future)
fcastProphet=forecast[['ds', 'yhat']].iloc[-len(test):]
fcastProphet.set_index('ds', inplace=True)
fcastProphet.columns=['Price_USD']
fcastProphet.index.freq='D'

mean_absolute_percentage_error(fcastProphet['Price_USD'], test['Price_USD'])
```

### SARIMA with Box-Cox

```python
from scipy import stats
x=df_NVidia['Price_USD']
xt,lambda_mle=stats.boxcox(x)
x_log=np.log(x)
auto_arima(xt,seasonal=True,m=7,trace=True).summary()
```

```python
train_box_cox=xt[:-30]
test_box_cox=xt[-30:]
from scipy.special import boxcox, inv_boxcox
model_box_cox=SARIMAX(train_box_cox,order=(0,1,0),seasonal_order=(0,0,0,7))
results_box_cox=model_box_cox.fit()
fcastSARIMA_box_cox=results_box_cox.predict(start=start, end=end, dynamic=False)

fcastSARIMA_BoxCox=pd.DataFrame(inv_boxcox(fcastSARIMA_box_cox, lambda_mle))
fcastSARIMA_BoxCox.set_index(test.index, inplace=True)
fcastSARIMA_BoxCox.colmuns=['FirstTimeVisit']
mean_absolute_percentage_error(inv_boxcox(fcastSARIMA_box_cox, lambda_mle), test['Price_USD'])
```
